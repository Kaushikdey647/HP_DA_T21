```python
def convertToJson(newfile):  #newfile is dataframe assuming column names are the keys
    res=[]
    mapping=[]
    for r in newfile:
        mapping.append(r)
    for i in range(newfile.shape[0]):
        dic={}
        for keys in mapping:
            dic[keys]=newfile.iloc[i][keys]
        res.append(dic)
    return res

ans=convertToJson(df)
for x in ans:
    x['Ticker']=x['Ticker'].upper()

new=[]

for i in range(178):
    dict={}
    for x in ans[i]:
        if x == 'Ticker':
            dict[x]=ans[i][x]
            continue
        else:
            orig=x
            res=x.split(' - ')
            if int(res[0])<10:
                res[0]='0'+res[0]
            x=res[1]+'-'+res[0]+'-01'
            dict[x]=(ans[i][orig])

    new.append(dict)
            
```
